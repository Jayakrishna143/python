##  String methods without using string methods using functions and return s	statements.


```python
def rindex(s: str, s1: str):
    for i in range(len(s) - len(s1), -1, -1):  # Loop backward
        if s[i:i + len(s1)] == s1:
            return i
    return -1
rindex("hello world", "z")
```

```python
# index
def index(s: str, s1: str):
    for i in range(len(s)):  # Loop backward
        if s[i:i + len(s1)] == s1:
            return i
    return -1
index('my name is jayakrishna','name')
```

```python
#rfind
def rfind(s: str, s1: str):
    for i in range(len(s)-len(s1)+1,-1,-1): #loop backward
        if s[i:i+len(s1)]==s1:
            return i
            break
    else:
        return -1
rfind('my name is jayakrihsna reddy','d')
```

```python
#startswith
def startswith(s: str, s1: str):
    if s[0:len(s1)]==s1:
        return True
    else:
        return False
startswith('my name is jayakrihsna reddy','my')
```

```python
#partition
def partition(s: str, s1: str):
    if s1 not in s:
        print(s,'','')
    else:
        for i in range(len(s)-len(s1)+1):
            if s[i:i+len(s1)]==s1:
                before=s[0:i]
                after=s[i+len(s1):]
                return before,s1,after
partition('my name is jayakrihsna reddy','name')
```

```python
def swapcase(s: str):
    s1=''
    for i in s:
        if ord(i) in range(ord('a'),ord('z')+1):
            s1 +=chr(ord(i)-32)
        elif ord(i) in range(ord('A'),ord('Z')+1):
            s1 +=chr(ord(i)+32)
        else:
            s1+=i
    return s1
swapcase('My Name Is Jayakrihsna Reddy')
```

```python
#count
def count(s: str, s1: str):
    c=0
    for i in range(len(s)-len(s1)+1):
        if s[i:i+len(s1)]==s1:
            c+=1
    return c
count('my name is jayakrihsna reddy','na')
```

```python
# lower
def lower(s: str):
    s1=''
    for i in s:
        if ord(i) in range(ord('A'),ord('Z')+1):
            s1+=chr(ord(i)+32)
        else:
            s1+=i
    return s1
lower('My Name Is Jayakrihsna Reddy')
```

```python
# is lower

def is_lower(s:str):
    x=  True
    for i in s:
        if ord(i) in range(ord('a'),ord('z')):
            x = True
        else:
            x= False
    return x
is_lower('jayakrihsna')
```

```python
#isalpha
def is_alpha(s:str):
    x=True
    for i in s:
        if ord(i) not in range(ord('a'),ord('z')) and ord(i) not in range(ord('A'),ord('Z')):
            x = False
    return x
is_alpha('Jayakrishna ')
```

```python
#isdecimal
def is_decimal(s:str):
    x=True
    for i in s:
        if ord(i) not in range(ord('0'),ord('9')):
            x=False
    return x
is_decimal('1234a')
```

```python
# zfill
def zfill(s:str , j:int):
    if j <= len(s):
        return s
    else:
        return f'0' * (j-len(s))+s
zfill('jaya',10)
```

```python
#capitalize string
def capitalize_string(s:str):
    result = ""
    for i in range(len(s)):
        if i == 0:
            if 'a' <= s[i] <= 'z':
                result += chr(ord(s[i]) - 32)
            else:
                result += s[i]
        else:
            result += s[i]
    return result
capitalize_string('jaya krihsna')
```

```python
# split
# s1 is dellimiter A character from which the string is split
def split_string(s: str, s1: str):
    result = []
    current_word = ""
    for i in range(len(s)):
        if s[i:i + len(s1)] == s1:
            result.append(current_word)
            current_word = ""
            i += len(s1) - 1
        else:
            current_word += s[i]
    result.append(current_word)
    return result
split_string('my name is jayakrihsna reddy',' ')
```

```python
# ljust
def ljust(s: str, width: int, char: str = "*"):
    if width <= len(s):
        return s
    result = s
    for _ in range(width - len(s)):
        result += char

    return result
ljust('jaya',10)
```

```python
#prefix
def remove_prefix(string:str,prefix:str):
    if len(string) >= len(prefix):
        match = True
        for i in range(len(prefix)):
            if string[i] != prefix[i]:
                match = False
    if match:
        return string[len(prefix):]
    else:
        return 'not found'
remove_prefix('jayakrihsna','z')
```

```python
#removesuffix
def remove_sufffix(string:str,suffix:str):
    if len(string) >= len(suffix):
        match = True
        for i in range(-len(suffix)):
            if string[i] != suffix[i]:
                match = False
    if match:
        return string[:-len(suffix)]
    else:
        return 'not found'
remove_sufffix('jayakrihsna','na')
```
