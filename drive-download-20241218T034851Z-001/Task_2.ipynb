# 2. Operators
Operators are used to perform operations on variables and values.

### Types of Operators
1.Arithmetic Operators: +, -, *, /, %, **, //

2.Comparison Operators: ==, !=, <, >, <=, >=

3.Logical Operators: and, or, not

4.Assignment Operators: =, +=, -=, *=, /=

```python
# Arithmetic and comparison
a = 45
b = 16
print(a + b, a - b, a * b, a / b, a // b, a % b)
print(a > b, a == b)

# Logical operators
print(a > 10 and b < 10)
print(not (a > b))
```

# Conditional statements
Conditional statements allow us to execute specific blocks of code based on certain conditions.
These conditions are evaluated as either True or False.

 1.Conditional statements use keywords like if, elif, and else.

2.If Statement: Executes a block of code if the condition is True.

3.Elif Statement: Allows us to check multiple conditions. It stands for "else if.

4.Else Statement: Executes a block of code if all before conditions are False.

5.Logical Operators: and, or, and not are used.

6.Nested Conditions: we can have conditional statements inside other conditional statements.

```python
#if statement
age = 22
if age >= 18:
    print("You are an adult.")
```

```python
#else,elif,if statements
age  = 55
if age >= 80:
    print("Bamma garu inka chanipoledha meru")
elif age >= 60:
    print("inkennalu untav bamma")
else:
    print("nuv bammavi enti inka sweet 16 aeethe ")
```

```python
#Using Logical Operators:
temperature = 25
if temperature > 20 and temperature < 30:
    print("The weather is pleasant.")
```

```python
#Nested Conditions:
num = 45
if num > 10:
    if num % 2 == 0:
        print("The number is even and greater than 10.")
    else:
        print("The number is odd and greater than 10.")
```

# 6. Lists

```python
# Creating and manipulating a list
fruits = ["apple", "banana", "cherry"]
fruits.append(1)
print(fruits)
print(fruits[1])
```

# 9. Loops
Loops allow you to iterate over a sequence (like a list or range).


For Loop Example:

```python


# Iterating over a list
for fruit in fruits:
    print(fruit)
```

While Loop Example:


```python
# Using a while loop
count = 0
while count < 5:
    print(count)
    count += 1
```

# The Evolution of Programming Languages
#### Introduction
Programming languages have undergone significant transformations since their inception. From early machine-level instructions to high-level, human-readable syntax, these languages have been at the heart of technological progress. This article explores the evolution of programming languages, their milestones, and their impact on computing and society.

### 1. The Early Era: Machine and Assembly Language
The first programming languages were tightly coupled to the hardware they controlled. Machine language, consisting of binary instructions, was the only way to program early computers. Writing code was laborious and prone to errors, as it required intimate knowledge of the underlying hardware.

### Machine Code:
Binary instructions directly executed by the CPU.
#### Example: 10101100 01011001

Extremely efficient but difficult for humans to write and debug.
### Assembly Language:
A slight abstraction above machine code, using mnemonics instead of binary.

Example
1. MOV A, 10
2. ADD A, B

Required an assembler to translate into machine code.

##2. The Birth of High-Level Languages (1950s-1960s)
High-level languages emerged to simplify programming by abstracting hardware complexities. This innovation made programming accessible to a wider audience.


1.   FORTRAN (1957): The first high-level language designed for scientific computation.

2.   COBOL (1959): Optimized for business applications with readable syntax.

3.   LISP (1958): Introduced for artificial intelligence research, featuring recursive functions and symbolic computation.









## 3. The Structured Programming Revolution (1970s)
The 1970s witnessed a shift toward structured programming, emphasizing modularity and clarity. This era saw the rise of block-structured languages like:


1.   C (1972): A versatile language balancing low-level access with high-level constructs.

2.  Pascal (1970): Promoted teaching structured programming with strong typing and clean syntax.

Key advancements:

*   Emphasis on modular programming.
*   Introduction of control structures (if-else, loops).


## 4. Object-Oriented Paradigm (1980s)
The 1980s introduced the object-oriented paradigm to improve code reuse and maintainability.

1.   malltalk: The pioneer of OOP concepts like classes and inheritance.
2.   C++ (1985): Extended C with object-oriented features.
3.   Java (1995): A platform-independent, OOP-focused language.


## 5. The Rise of Scripting and Web Development (1990s)
As the internet gained popularity, languages adapted to support web development and rapid prototyping.

1.   Python (1991): Known for simplicity and readability.

2.   JavaScript (1995): Revolutionized web interactivity

3. PHP (1995): Became the backbone of server-side scripting.





## 6. Modern Trends and Multi-Paradigm Languages (2000s-Present)
In recent decades, languages have become more versatile, blending multiple paradigms to suit various needs.


1.   Ruby (1995): Emphasized developer happiness with clean syntax
2.  Go (2009): Focused on simplicity and performance for concurrent programming.
3.   Rust (2010): Prioritized safety and concurrency without compromising performance.
4.   Kotlin (2011): A modern alternative to Java for Android development.






## 7. The Future of Programming Languages
The evolution of programming languages continues, driven by trends like:


*  Machine Learning and AI: Languages like Python dominate due to their rich ecosystems.
*   Low-code/No-code platforms: Democratizing software development.
*  Quantum Computing: Languages like Q# and Cirq are emerging.
